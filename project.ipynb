## Исследование задачи

### Баланс классов

```python
# посмотрим, сколько ушедших клиентов в датасете

data.pivot_table(index = 'exited', values = 'customer_id', aggfunc = 'count', margins=True)
```

Около 20% от всех записей относятся к ушедшим клиентом. Таким образом, классы в целевом признаке сильно несбалансированы.

Обучим модель на текущих данных.

### Деление выборки на сеты

```python
# разделим выборку на обучающие признаки и целевой параметр.

features = data_work.drop('exited', axis=1)
target = data['exited']
```

```python
# разделим выборки на обучающую, валидационную и тестовую в 2 этапа.

# обучающая и тестовая:

features_train, features_test, target_train, target_test = train_test_split (features, target, test_size = 0.20, random_state = 12345)
```

```python
# разделим обучающую выборку на 2 части: для обучения и для валидации результатов

features_train, features_valid, target_train, target_valid = train_test_split (features_train, target_train, test_size = 0.25, random_state = 12345)
```

Сделаем масштабирование количественных признаков.

```python
numeric = ['credit_score', 'age', 'tenure', 'balance', 'num_of_products', 'estimated_salary']

scaler = StandardScaler()
scaler.fit(features_train[numeric])
features_train[numeric] = scaler.transform(features_train[numeric])
display(features_train.head())
```

```python
# применим обученный скелер к валидационной выборке
features_valid[numeric] = scaler.transform(features_valid[numeric])
display(features_valid.head())
```

```python
# применим обученный скелер к тестовой выборке
features_test[numeric] = scaler.transform(features_test[numeric])
display(features_test.head())
```

### Выбор модели

```python
# найдем наилучшие параметры для Дерева решений

best_score = 0
best_depth = 0

for depth in range (1, 10, 1):
    model = DecisionTreeClassifier(max_depth = depth, random_state = 12345)
    model.fit(features_train, target_train)
    predictions = model.predict(features_valid)
    f1 = f1_score(target_valid, predictions)
    if f1 > best_score:
        best_score = f1
        best_depth = depth

print ('Значение F1 меры:', round(best_score,3))
print ('Максимальна глубина для модели дерева решений:', best_depth)

probabilities = model.predict_proba(features_valid)
probabilities_one_valid = probabilities[:, 1]
auc_roc = roc_auc_score(target_valid, probabilities_one_valid)
print('Значение AUC_ROC', round(auc_roc,3))
```

```python
# найдем оптимальное количество деревьев и глубину в модели Случайного леса

best_score = 0
estimator_num = 0
best_depth = 0

for estimator in range (1, 50, 1):
    for depth in range (1,10):
        model = RandomForestClassifier(max_depth = depth, n_estimators = estimator, random_state = 12345)
        model.fit(features_train, target_train)
        predictions = model.predict(features_valid)
        f1 = f1_score(target_valid, predictions)
        if f1 > best_score:
            best_score = f1
            estimator_num = estimator
            best_depth = depth

print ('Значение F1 меры:', round(best_score, 3))
print ('Оптимальное кол-во деревьев:', estimator_num)
print ('Оптимальная глубина:', best_depth)

probabilities = model.predict_proba(features_valid)
probabilities_one_valid = probabilities[:, 1]
auc_roc = roc_auc_score(target_valid, probabilities_one_valid)
print('Значение AUC_ROC', round(auc_roc, 3))
```

```python
# Обучим модель регрессии

model = LogisticRegression(random_state = 12345)
model.fit(features_train, target_train)
predictions = model.predict(features_valid)
f1 = f1_score(target_valid, predictions)

print ('Значение F1 меры модели логистической регрессии:', round(f1,3))

probabilities = model.predict_proba(features_valid)
probabilities_one_valid = probabilities[:, 1]
auc_roc = roc_auc_score(target_valid, probabilities_one_valid)
print('Значение AUC_ROC', round(auc_roc,3))
```

Наилучшей моделью оказалась модель Случайного леса.

### Модели с методом увеличения выборки для меньшего класса

```python
# попробуем обучить модель методом увеличения выборки
# напишем функцию для увеличения выборки

def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

# применим ее
features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
```

Подберем лучшую модель:

```python
# модель Решаюшего дерева

best_score = 0
best_depth = 0

for depth in range (1, 10, 1):
    model = DecisionTreeClassifier(max_depth = depth, random_state = 12345)
    model.fit(features_upsampled, target_upsampled)
    predictions = model.predict(features_valid)
    f1 = f1_score(target_valid, predictions)
    if f1 > best_score:
        best_score = f1
        best_depth = depth

print ('Значение F1 меры:', round(best_score,3))
print ('Максимальна глубина для модели дерева решений:', best_depth)

probabilities = model.predict_proba(features_valid)
probabilities_one_valid = probabilities[:, 1]
auc_roc = roc_auc_score(target_valid, probabilities_one_valid)
print('Значение AUC_ROC', round(auc_roc,3))
```

```python
# модель Случайного леса

best_score = 0
estimator_num = 0
best_depth = 0

for estimator in range (1, 50, 1):
    for depth in range (1,10):
        model = RandomForestClassifier(
            max_depth = depth, n_estimators = estimator, 
            random_state = 12345
        )
        model.fit(features_upsampled, target_upsampled)
        predictions = model.predict(features_valid)
        f1 = f1_score(target_valid, predictions)
        if f1 > best_score:
            best_score = f1
            estimator_num = estimator
            best_depth = depth

print ('Значение F1 меры:', round(best_score, 3))
print ('Оптимальное кол-во деревьев:', estimator_num)
print ('Оптимальная глубина:', best_depth)

probabilities = model.predict_proba(features_valid)
probabilities_one_valid = probabilities[:, 1]
auc_roc = roc_auc_score(target_valid, probabilities_one_valid)
print('Значение AUC_ROC', round(auc_roc, 3))
```

### Модели с методом уменьшения выборки для большего класса

```python
# попробуем обучить модель методом уменьшения выборки
# напишем функцию

def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    features_downsampled = pd.concat([features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat([target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

# применим ее
features_downsampled, target_downsampled = downsample(features_train, target_train, 0.25)
```
